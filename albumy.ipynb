```python
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
```

```python
df = data[0]
```

```python
df.rename(columns={'TITLE':'TYTUŁ','ARTIST':'ARTYSTA','YEAR':'ROK','HIGH POSN':'MAX POZ'}, inplace=True)
```

```python
artist = df['ARTYSTA'].nunique()
```

```python
print(f"Na liście znajduję się {artist} pojedynczych artystów.")
```

```python
df['ARTYSTA'].value_counts()
```

```python
most_popular_artist_1 = df['ARTYSTA'].value_counts().index[0]
```

```python
most_popular_artist_2 = df['ARTYSTA'].value_counts().index[1]
```

```python
print(f"Na liście najcześciej znajdowali się: {most_popular_artist_1} i {most_popular_artist_2} ")
```

```python
df.rename(str.capitalize, axis='columns', inplace=True)
```

```python
df.drop('Max poz', axis=1, inplace=True)
```

```python
df['Rok'].value_counts()
```

```python
top_year_1 = df['Rok'].value_counts().index[0]
```

```python
top_year_2 = df['Rok'].value_counts().index[1]
```

```python
print(f"W roku {top_year_1} i {top_year_2} wyszło najwięcej albumów znajdujących się na liście.")
```

```python
df[(df['Rok']>1960)&(df['Rok']<=1990)]
```

```python
quantity_of_albums = len(df[(df['Rok']>1960)&(df['Rok']<=1990)])
```

```python
print(f"Lista zawiera {quantity_of_albums} albumy wydane między rokiem 1960 a 1990.")
```

```python
latest_album = df['Rok'].max()
```

```python
print(f"Najmłodszy album został wydany w roku {latest_album}")
```

```python
artist_list = df.groupby('Artysta').agg('Rok').min()
```
